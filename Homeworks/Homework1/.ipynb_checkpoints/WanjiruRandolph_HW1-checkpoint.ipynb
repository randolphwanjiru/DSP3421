```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#losding the dataset
df = pd.read_csv('/movie.csv')

#1. Data types and summary
data_types_summary = df.dtypes
print("1. Data types and summary:")
print(data_types_summary)
print("\n Summary:")
print(summary)


#2. descriptive statistics
descriptive_statistics = df.describe()
print("\n2. Descriptive statistics")
print(descriptive_statistics)


#3. Disrtibution Anlaysis
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')

    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[column])
    plt.title(f'Boxplot of {column}')

    plt.show()


# 4. Variability Analysis
highest_variability_column = df.var(numeric_only=True).idxmax()
print("\n4. Variability Analysis:")
print(f"The independent variable with the highest variability is: {highest_variability_column}")

# 5. Correlation Analysis
# Check the column names
print("\nColumn Names:")
print(df.columns)

# 5. Correlation Analysis
# Check the column name for 'Sales Revenue'
correlation_with_sales = df.corr()['Sales Revenue_M'].sort_values(ascending=False)
strong_correlation_variables = correlation_with_sales[abs(correlation_with_sales) > 0.5]
print("\n5. Correlation Analysis:")
print(f"Variables with strong correlation with Sales Revenue:\n{strong_correlation_variables}")


# 6. Average Profit by Genre
average_revenue_by_genre = df.groupby('Genre')['Sales Revenue_M'].mean().idxmax()
print("\n6. Average Profit by Genre:")
print(f"The genre with the highest average sales revenue is: {average_revenue_by_genre}")


# 7. Marketing Spend Analysis
budget_vs_revenue_corr = df['Budget_M'].corr(df['Sales Revenue_M'])
print("\n7. Marketing Spend Analysis:")
print(f"Correlation between Budget and Sales Revenue: {budget_vs_revenue_corr}")

# 8. Pairwise Comparison
sns.pairplot(df)
plt.show()

# 9. Missing Values Investigation
missing_values_summary = df.isnull().sum()
print("\n9. Missing Values Investigation:")
print(missing_values_summary)

# 10. Noisy Data Analysis
# Examine the dataset for potential outliers using boxplots
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()

# 11. Data Deduplication
deduplicated_df = df.drop_duplicates()
print("\n11. Data Deduplication:")
print(f"Number of duplicate records removed: {len(df) - len(deduplicated_df)}")



```
